B. Interpole el siguiente conjunto de datos:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos proporcionados
xs = np.array([
    0.0003, 0.0822, 0.2770, 0.4212, 0.4403, 0.5588, 0.5943,
    0.6134, 0.9070, 1.0367, 1.1903, 1.2511, 1.2519, 1.2576,
    1.6165, 1.6761, 2.0114, 2.0557, 2.1610, 2.6344
])

ys = np.array([
    1.1017, 1.5021, 0.3844, 1.3251, 1.7206, 1.9453, 0.3894,
    0.3328, 1.2887, 3.1239, 2.1778, 3.1078, 4.1856, 3.3640,
    6.0330, 5.8088, 10.5890, 11.5865, 11.8221, 26.5077
])

# Grado del polinomio (puedes probar con 2, 3, 4, etc.)
grado = 3

# Ajuste del polinomio por mínimos cuadrados
coeficientes = np.polyfit(xs, ys, grado)
polinomio = np.poly1d(coeficientes)

# Evaluar el polinomio en puntos intermedios para graficar
x_interp = np.linspace(min(xs), max(xs), 500)
y_interp = polinomio(x_interp)

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(xs, ys, 'o', label='Datos originales')
plt.plot(x_interp, y_interp, '-', label=f'Polinomio grado {grado}')
plt.title('Interpolación por Mínimos Cuadrados')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Mostrar coeficientes
print(f'Coeficientes del polinomio de grado {grado}:')
print(polinomio)
```
